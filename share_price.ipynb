```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Fetch Jindal Steel stock data (NSE: JINDALSTEL)
ticker = "JINDALSTEL.NS"  # Use "JINDALSTEL.BO" for BSE
stock = yf.Ticker(ticker)

# Fetch data from 2025 to 2030
data = stock.history(start="2025-01-01", end="2030-12-31")

# Display the first few rows
print(data.head())

# Plot the stock price trend
plt.figure(figsize=(12, 5))
plt.plot(data.index, data['Close'], label="Close Price")
plt.title("Jindal Steel Share Price (2025-2030)")
plt.xlabel("Year")
plt.ylabel("Stock Price (INR)")
plt.legend()
plt.show()

```
